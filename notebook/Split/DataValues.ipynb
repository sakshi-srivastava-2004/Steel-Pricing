```python
import openpyxl
import pandas as pd

# Load workbook
file_path = "rebar_dom.xls"
wb = openpyxl.load_workbook(file_path, data_only=True)
target_color = "FFFFFF00"  # Yellow fill in RGB

# Dictionary to store data from each sheet
sheet_data = {}

# Loop through all sheet names
for sheet_name in wb.sheetnames:
    ws = wb[sheet_name]
    highlighted_headers = []

    for col in ws.iter_cols(min_row=1, max_row=2):  # First two rows
        top_cell = col[0]   # Row 1: Highlight check
        header_cell = col[1]  # Row 2: Actual header

        # Get fill color
        fill = top_cell.fill
        color = fill.start_color.rgb if fill.start_color.type == 'rgb' else None

        # Match yellow color
        if color == target_color:
            highlighted_headers.append(header_cell.value)

    if highlighted_headers:
        # Read only selected columns from this sheet
        df = pd.read_excel(file_path, sheet_name=sheet_name, header=1, usecols=highlighted_headers)
        sheet_data[sheet_name] = df
        print(f"Processed sheet: {sheet_name}, columns: {highlighted_headers}")
    else:
        print(f"No highlighted columns found in sheet: {sheet_name}")

# Example: Access a DataFrame from a sheet
# df_math = sheet_data['Math']
```
